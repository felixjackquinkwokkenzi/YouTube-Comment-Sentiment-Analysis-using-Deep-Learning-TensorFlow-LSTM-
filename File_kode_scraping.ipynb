```python
import pandas as pd
from googleapiclient.discovery import build


# Scraping data komentar youtube
api_key = 'AIzaSyCwasRLx_RO55IuQk5vWeS_E1lWkJth_vw'
youtube = build('youtube', 'v3', developerKey=api_key)

def get_video_comments(video_id, max_results):
    comments = []
    next_page_token = None

    while len(comments) < max_results:
        request = youtube.commentThreads().list(
            part='snippet',
            videoId=video_id,
            maxResults=min(100, max_results - len(comments)),
            pageToken=next_page_token
        )
        response = request.execute()

        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']['textOriginal']
            comments.append(comment)

        next_page_token = response.get('nextPageToken')
        if not next_page_token:
            break

    return comments

# Gantilah 'VIDEO_ID' dengan ID video yang ingin Anda scrape
video_id = 'https://www.youtube.com/watch?v=_uQrJ0TkZlc'[-11:]
comments = get_video_comments(video_id, 10000)

df = pd.DataFrame(comments, columns=['komentar'])
df.to_csv('youtube_comments.csv', index=False)
```
